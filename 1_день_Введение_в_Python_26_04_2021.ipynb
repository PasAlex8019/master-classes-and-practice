````
import pandas as pd
````

````
data = pd.read_excel('weather.xls', skiprows=6)  # читаем файл, отбросив первые 6 строк с комментариями
````

````
data['dates'] = pd.to_datetime(data['Местное время в Москве (ВДНХ)'], dayfirst=True)  # Преобразование дат в русском формате в формат, пригодный для языка Python
````

````
data.head(10)
````

````
x = data['dates']
y = data['T']
plt.plot(x, y)
````

````
# Фильтрация данных по условию
day_end = pd.Timestamp(year=2017, month=3, day=1)
condition = data['dates'] < day_end

data[condition]
````

````
data_short = data[condition]

plt.figure(figsize=[20, 5])
x = data_short['dates']
y = data_short['T']
plt.plot(x, y)
````
