**1.b.** Постройте следующий график:

- процентое изменение цены за акцию Tesla за день по отношению к цене на открытии торгов.

На оси **x** должны быть отмечены даты из data.index. Значения оси **y** должны отражать, на сколько процентов увеличилась (положительное число) или на сколько процентов упала (отрицательное число) цена акции на закрытии Close по сравнению с ценой открытия Open. 

В какой день наблюдалось самое резкое понижение цены за акцию? А самое большое повышение? На сколько процентов?

*Подсказка: создайте новую колонку Diff, в которой посчитайте изменение акции за день в процентах*

**Ваш ответ:**

```python
#data.head()
```

```python
df_2  = yf.download('TSLA')
df_2['Diff'] = (df_2.Close - df_2.Open) * 100 / df_2.Open
df_2.head()
```

```python
fig_2 = px.line(df_2, x = df_2.index, y = 'Diff',\
                title='График: процентое изменение цены за акцию Tesla за день по отношению к цене на открытии торгов.')
                 
fig_2.show()
```

```python
df_2_min = df_2[df_2['Diff']== df_2.Diff.min()]
df_2_max = df_2[df_2['Diff']== df_2.Diff.max()]

#df_2_min.Diff[0]
print(f'День в котором наблюдалось самое резкое понижение цены за акцию : {df_2_min.index.day_name()[0]}, на {round(df_2_min.Diff[0], 2)} процентов ')
print(f'День в котором наблюдалось самое большое повышение цены за акцию : {df_2_max.index.day_name()[0]}, на {round(df_2_max.Diff[0], 2)} процентов ')
```

Напишите месяц, когда объем акций впервые превысил 60M. Какая средняя цена акции на закрытии торгов соответствует этому месяцу?

*Подсказка: используйте параметр hover_data.*

**Ваш ответ:**

```python
df_3 = df_2.copy()
df_3.head()
```

```python
df_3.index = df_3.index.to_period('M')
df_3.head()
```

```python
df_3_mean = df_3.groupby(df_3.index).agg('mean')
df_3_mean.index = df_3_mean.index.to_timestamp()
df_3_mean.info()
```

```python
fig_3 = px.line(df_3_mean, x = df_3_mean.index, y = 'Volume', height = 400, width = 1500, \
                hover_data = ['Close'],\
                title='График: изменения объемов продаж акций Tesla.')
                 
fig_3.show()
```

```python
max_df_3 = df_3_mean[df_3_mean.Volume > 60000000].reset_index()
max_df_3['month_name'] = max_df_3.Date.dt.month_name()
max_df_3 = max_df_3.loc[max_df_3.index[0]][['month_name', 'Close']]
max_df_3
```

```python
# При выходе бренда на рынок общие объемы начали падать, с объема превышающего 60 М,
#если рассматривать  рост после этого падение и возврат к прежним показателям, то это бует второе камание данного уровня
print(f'Месяц, когда объем акций впервые превысил 60M - {max_df_3.month_name}.\n\
Средняя цена акции на закрытии торгов соответствующая этому месяцу - {round(max_df_3.Close, 2)}')
```

```python
# Превое превышение щбъема акций 60 М, после провала при открытии
max_df_second = df_3_mean[(df_3_mean.Volume > 60000000) & (df_3_mean.index > '2012')].reset_index()
max_df_second['month_name'] = max_df_second.Date.dt.month_name()
max_df_second = max_df_second.loc[max_df_second.index[0]][['month_name', 'Close']]
#max_df_second
print(f'Месяц, когда объем акций впервые после обвала  превысил 60M - {max_df_second.month_name}.\n\
Средняя цена акции на закрытии торгов соответствующая этому месяцу - {round(max_df_second.Close, 2)}')
```
