```python
import re
corpus = [] # сщздаем объект list в переменной corpus
regular_expr = r'.+?(?=,)' # регулярное выражение для поиска слов в тексте и их извлечения
reg_expr_compiled = re.compile(regular_expr) # производим компеляцию регулярного выражения regular_expr
# для его применения к тексту

# создаем цикл for и запускаем по всем строкам текста
for raw_text in g.Description_list.values: 
    # приводим к нижнему регистру
    #raw_text_lower = raw_text.lower()
    # разбиваем текст на отдельные слова с помощью регулярных выражений reg_expr_compiled
    text_by_words = reg_expr_compiled.findall(raw_text)
    # после преобразования текста в последовательность слов, сохраняем ее в список corpus
    corpus.append(text_by_words)
corpus

```

```python
# полученный список нормализованных слов преобразуем в DataFrame
doc_count = len(corpus) # считаем колличество всех всех текстовых документов
# т.е. колличество всех текстовых документов которые входят в него
# формируем дополнительно два докумкнта-списка это будут колонки нашего дата фрейма
doc_ids = [] # id наших строк(content => 4)
tokens = [] # сырые обработанные-нормализованные слова
# запускаем цикл for который проходит по всем документам
for doc_id in range(doc_count):
    # внутри каждого документа через внутрений for обходим все токены
    for token in corpus[doc_id]:
        # и прекрепляем к листу doc_ids => id-шники документа
        doc_ids.append(doc_id)
        # а к листу tokens => token докуменда
        tokens.append(token)
```

```python
tokens_df = pd.DataFrame({
    'doc_id':doc_ids,
    'Description':tokens
})

tokens_df
```

```python
tokens_df = tokens_df.assign(dummy = 1)

tokens_df
```
